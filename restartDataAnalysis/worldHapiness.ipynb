```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr,
            annot=True,        # hücrelerdeki korelasyon katsayılarını yaz
            fmt=".2f",         # iki ondalık basamak
            cmap="coolwarm",   # mavi–kırmızı renk paleti
            center=0,          # 0 ortalanır (nötr renk)
            linewidths=0.5,    # hücre arası ince çizgiler
            square=True)       # kare hücreler
plt.title("Korelasyon Matrisi (Heatmap)", fontsize=14, pad=15)
plt.show()
```

```python
sns.pairplot(df, vars=corr.columns, diag_kind="kde", corner=True)
plt.suptitle("Korelasyonların Dağılım Görselleştirmesi", y=1.02)
plt.show()
```

```python
# why Australia and New Zealand have highest happiness score
oceania = df[df["Region"].isin(["Australia and New Zealand", "Oceania"])]
region_avg = df.groupby("Region")["Happiness Score"].mean().sort_values(ascending=False)
region_avg.plot(kind="bar", color="skyblue", figsize=(10,5), title="Average Happiness Score by Region")
```

```python
factors = ["Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", 'Freedom', "Trust (Government Corruption)", "Generosity","Dystopia Residual"]
oceania_mean = df[df["Country"].isin(["Australia","New Zealand"])][factors].mean()
global_mean = df[factors].mean()

comparison = pd.DataFrame({
    "Oceania Mean": oceania_mean,
    "Global Mean": global_mean,
    "Difference": oceania_mean - global_mean
}).sort_values("Difference", ascending=False)
comparison
```

```python
corr = df[factors + ["Happiness Score"]].corr()["Happiness Score"].sort_values(ascending=False)
sns.barplot(x=corr.values, y=corr.index, palette="coolwarm")
plt.title("CORR WITH HAPPINESS SCORE")
plt.xlabel("CORRELATION")
```

```python
import statsmodels.api as sm

X = df[factors]
y = df["Happiness Score"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

Are the variables that increase happiness scores in Western Europe, North America, and New Zealand–Australia the same in terms of importance and impact?

```python
regions = {
    "Western Europe": "Western Europe",
    "North America": "North America",
    "Oceania": "Australia and New Zealand"
}
subset = df[df["Region"].isin(regions.values())]
```

```python
region_corr = (subset.groupby("Region")[factors + ["Happiness Score"]]
               .corrwith(subset["Happiness Score"])
               .unstack(level=0))
```

```python
region_corr.T.plot(kind="bar", figsize=(10,6))
plt.title("HAPPINESS CORRELATIONS BY REGION")
plt.ylabel("CORRELATION")
plt.grid(True, alpha=0.3)
plt.show()
```
