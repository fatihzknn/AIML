```python
pop_cols = [c for c in df.columns if "Population" in c and c[:4].isdigit()]

df_long = df.melt(
    id_vars=["Country", "Continent"],
    value_vars=pop_cols,
    var_name="Year",
    value_name="Population"
)
# "2022 Population"  sadece yl
df_long["Year"] = df_long["Year"].str.extract(r"(\d{4})").astype(int)
```

```python
df_long
```

```python
df_long = df_long.sort_values(["Country", "Year"])
df_long
```

```python
df_long["growth"] = df_long.groupby("Country")["Population"].pct_change()
avg_growth = df_long.groupby("Country")["growth"].mean()
avg_growth
```

```python
def estimate_r_loglin(years, pops):
    mask = np.isfinite(pops)
    if mask.sum() < 2:
        return np.nan
    # log(P) = a + b*year    yllk byme ~ exp(b) - 1
    b, a = np.polyfit(years[mask], np.log(pops[mask]), 1)
    return np.exp(b) - 1
# stee bal: lke iin r bulunamazsa kullanlacak global/kita medyan
def compute_fallback_r(df_long):
    g = df_long.copy()
    g["Population"] = pd.to_numeric(g["Population"], errors="coerce")
    g = g.sort_values(["Country","Year"])
    g["growth"] = g.groupby("Country")["Population"].pct_change()
    return g["growth"].median(skipna=True)  # global medyan
# fallback_r = compute_fallback_r(df_long)

def fill_pop(group, fallback_r=0.0):
    gr = group.sort_values("Year").copy()
    # saysal tip garanti
    gr["Population"] = pd.to_numeric(gr["Population"], errors="coerce")

    years = gr["Year"].to_numpy()
    pops  = gr["Population"].to_numpy()
    valid = ~np.isnan(pops)

    # r: lkeye zel ortalama byme (NaN'leri d)
    g = pd.Series(pops, index=years).pct_change()
    r = g.mean(skipna=True)

    # r NaN ise fallback kullan
    if not np.isfinite(r):
        r = fallback_r  # rn. global/kta medyan veya 0.0

    # LER doru doldurma: P_t = P_prev * (1+r) ** delta_years
    for i in range(len(pops)):
        if np.isnan(pops[i]):
            # en yakn NCEK dolu deeri bul
            prev_idx = np.where(valid[:i])[0]
            if prev_idx.size > 0:
                j = prev_idx[-1]
                delta = int(years[i] - years[j])
                pops[i] = pops[j] * ((1.0 + r) ** max(delta, 1))
                valid[i] = True

    # GERYE doru doldurma: P_prev = P_next / (1+r) ** delta_years
    for i in range(len(pops)-1, -1, -1):
        if np.isnan(pops[i]):
            next_idx = np.where(valid[i+1:])[0]
            if next_idx.size > 0:
                k = i + 1 + next_idx[0]
                delta = int(years[k] - years[i])
                pops[i] = pops[k] / ((1.0 + r) ** max(delta, 1))
                valid[i] = True

    gr["Population"] = pops
    return gr
```

```python
# long formatta olmal: [Country, Continent, Year, Population]
# (Population saysal deilse nce to_numeric)
fallback_r = compute_fallback_r(df_long)  # veya kta bazl median growth
df_filled = df_long.groupby("Country", group_keys=False).apply(
    lambda g: fill_pop(g, fallback_r=fallback_r)
)
```

```python
df_filled
```

```python

country = "Turkey"

df_country = df_filled[df_filled["Country"] == country].sort_values("Year")

plt.figure(figsize=(8,5))
plt.plot(df_country["Year"], df_country["Population"]/1e6, marker='o', linestyle='-', color='steelblue')
plt.title(f"{country}  Nfus Deiimi (19702022)")
plt.xlabel("Yl")
plt.ylabel("Nfus (milyon)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
countries = ["Turkey", "Germany", "France"]

plt.figure(figsize=(9,6))
for c in countries:
    data = df_filled[df_filled["Country"] == c].sort_values("Year")
    plt.plot(data["Year"], data["Population"]/1e6, marker='o', label=c)

plt.title("lkelerin Nfus Deiimi (19702022)")
plt.xlabel("Yl")
plt.ylabel("Nfus (milyon)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_filled.loc[df_filled["Country"]=="Turkey", "Year"],
         df_filled.loc[df_filled["Country"]=="Turkey", "growth"]*100,
         marker='o', color='darkorange')
plt.title("Trkiye  Yllk Ortalama Nfus Bymesi (%)")
plt.xlabel("Yl")
plt.ylabel("Byme (%)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
countries = ["Turkey", "Germany", "France", "India"]
fig, axes = plt.subplots(len(countries), 1, figsize=(8,10), sharex=True)
for ax, c in zip(axes, countries):
    d = df_filled[df_filled["Country"]==c]
    ax.plot(d["Year"], d["Population"]/1e6, marker='o')
    ax.set_title(c)
    ax.grid(True, alpha=0.3)
plt.xlabel("Yl")
plt.tight_layout()
plt.show()
```
